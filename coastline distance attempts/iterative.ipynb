```python
from global_land_mask import globe
import pandas as pd
import geopy
import numpy as np
from geopy.distance import geodesic

sites = pd.read_csv(r"C:\Users\CAMG038492\Code\Climatology\archive\locations.csv")
mask = globe._mask

latitudes = np.linspace(-90, 90, mask.shape[0])
longitudes = np.linspace(-180, 180, mask.shape[1])
```

```python
testsite = sites.iloc[2]
```

```python
# determine if site is already water
land = globe.is_land(lat = testsite.Y, lon = testsite.X)
```

```python
coast_index = np.argwhere(mask)
```

```python
min = float('inf')

for i, j in coast_index:
    coast_lat = latitudes[i]
    coast_lon = longitudes[i]
    if globe.is_ocean(lat = coast_lat, lon = coast_lon):
        distance = geodesic((testsite.Y, testsite.X), (coast_lat, coast_lon)).kilometers

        if distance < min:
            min = distance

print(f"nearest coastline: {min:.2f} km")
```
